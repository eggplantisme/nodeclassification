```python
epsilons, train_ratios, accus = nodeAttrInfer.main1_symmetric_sbm(times=10)
```

```python
fig = plt.figure(figsize=(16, 16))
widths = [4]
heights = [4]
spec5 = fig.add_gridspec(ncols=1, nrows=1, width_ratios=widths, height_ratios=heights)
row = 0
col = 0
ax = fig.add_subplot(spec5[row, col])

ylabel = [round(e, 3) for e in epsilons]
xlabel = [round(r, 3) for r in train_ratios]
meanAccuData = []
for e in epsilons:
    temp = []
    for r in train_ratios:
        temp.append(np.mean(accus[e][r]))
    meanAccuData.append(temp)
print(meanAccuData)
ax.set_ylabel("epsilon")
ax.set_xlabel("train ratio")
ax.set_xticks(range(len(xlabel)))
ax.set_xticklabels(xlabel)
ax.set_yticks(range(len(ylabel)))
ax.set_yticklabels(ylabel)
im = ax.imshow(meanAccuData)
plt.colorbar(im)
```
