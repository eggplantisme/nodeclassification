```python
exp = empiric.EmpiricalExp('highSchool', load_hier=True, BHregularizer="Zeta_1")
fig = plt.figure(figsize=(16, 16))
widths = [4, 4]
heights = [4, 4]
spec5 = fig.add_gridspec(ncols=2, nrows=2, width_ratios=widths, height_ratios=heights)
row = 0
col = 0
ax = fig.add_subplot(spec5[row, col])
plt.spy(exp.Ag, markersize=1, rasterized=True, color=color0)
plt.xticks([])
plt.yticks([])
plt.xlabel('Adjacency matrix')

# calc first 2 eigenvalue in diff r 
d = np.sum(exp.Ag, 1)
rho = np.sum(d**2) / np.sum(d)
border = np.sqrt(rho)
print("rho", rho, "border", border)

rs = list(range(1, int(border))) + [border] + list(range(int(border), int(rho)+5))
x1, x2 = [], []
for r in rs:
    D = np.diag(d)
    B = np.eye(np.shape(exp.Ag)[0]) * (r**2 - 1) - r * exp.Ag + D
    w, _ = np.linalg.eig(B)
    w = np.sort(w)
    w = np.around(w, decimals=3)
    x1.append(w[0])
    x2.append(w[1])
print("length of rs", len(rs))
row = 0
col = 1
ax = fig.add_subplot(spec5[row, col])
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
ax.scatter(rs, x1, label='1')
ax.scatter(rs, x2, label='2')
ax.set_xlabel('First 2 Eigenvalue')
ax.set_aspect(0.1)
plt.axvline(border, color='k', lw=2, label=r"$\sqrt{\rho}$")
ax.legend()

BH = BetheHessian(exp.Ag, regularizer='Zeta_2')
zeta = BH.r
rs = list(range(-2, 3)) + [zeta, -zeta]
x1, x2, x3, x4, x5, x6, x7 = [], [], [], [], [], [], []
for r in rs:
    D = np.diag(d)
    B = np.eye(np.shape(exp.Ag)[0]) * (r**2 - 1) - r * exp.Ag + D
    w, _ = np.linalg.eig(B)
    w = np.sort(w)
    w = np.around(w, decimals=3)
    x1.append(w[0])
    x2.append(w[1])
    x3.append(w[2])
    x4.append(w[3])
    x5.append(w[4])
    x6.append(w[5])
    x7.append(w[7])
row = 1
col = 0
ax = fig.add_subplot(spec5[row, col])
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
ax.scatter(rs, x1, label='1')
ax.scatter(rs, x2, label='2')
ax.scatter(rs, x3, label='3')
ax.scatter(rs, x4, label='4')
ax.scatter(rs, x5, label='5')
ax.scatter(rs, x6, label='6')
ax.scatter(rs, x7, label='7')
ax.set_xlabel('First 7 Eigenvalue')
plt.axvline(border, color='k', lw=2, label=r"$\sqrt{\rho}$")
ax.legend()
```
